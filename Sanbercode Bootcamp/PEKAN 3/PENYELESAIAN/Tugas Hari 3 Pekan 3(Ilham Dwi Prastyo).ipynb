```python
import matplotlib.pyplot as plt
import numpy as np
```

<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 200)
tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
# code here
fig, ax = plt.subplots(nrows=1, ncols=3, sharex=True, sharey=True, squeeze=False)
ax[0][0].plot(x, tan, color='b', label='tanX')
ax[0][1].plot(x, cos, color='r', label='sinX')
ax[0][2].plot(x, sin, color='g', label='cosX')
ax[0][0].set_title('Grafik tanx')
ax[0][1].set_title('Grafik sinx')
ax[0][2].set_title('Grafik cosx')

fig.set_size_inches(20,8.2)
fig.legend(loc='upper center')

plt.show()
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
# code here
fig, ax = plt.subplots(figsize=(13,8.1))
ax.plot(x,y, lw=1.3)
ax.set_title("Grafik CosX",fontsize='medium')
ins1 = ax.inset_axes([-0.1,-0.1,0.3,0.3])
ins1.plot(x,y, color='b', lw=1.3)
ins1.plot(x,y3, color='r', lw=1.2)
ins1.set_title('Grafik CosX^2',fontsize='medium')
ins2 = ax.inset_axes([0.9,-0.1,0.3,0.3])
ins2.plot(x,y, color='b', lw=1.3)
ins2.plot(x,y3, color='g', lw=1.2)
ins2.set_title('Grafik CosX^3',fontsize='medium')
ins3 = ax.inset_axes([0.9,0.9,0.3,0.3])
ins3.plot(x,y, color='b', lw=1.3)
ins3.plot(x,y4, color='k', lw=1.2)
ins3.set_title('Grafik CosX^4',fontsize='medium')
ins4 = ax.inset_axes([-0.1,0.9,0.3,0.3])
ins4.plot(x,y, color='b', lw=1.3)
ins4.plot(x,y3, color='y', lw=1.2)
ins4.set_title('Grafik CosX^5',fontsize='medium')
plt.show()
```
