<h2>Soal 1: Pemahaman Component Matplotlib</h2>

- Jelaskan apa itu Figure
- Jelaskan apa itu Axis

jawab:

Figure, adalah window atau page atau halaman dalam objek visual. kalau kita ngegambar di kertas, maka kertas tersebutlah yang di namakan figure. Perintah kodenya yaitu: fig = plt.figure().<br>
Axis adalah suatu area di dalam figure dimana data akan di plot. Dalam Axis juga terdapat berbagai macam komponen lagi seperti x-axis, y-axis, dan lain sebagainya. Perintah kodenya yaitu: ax = fig.add_subplot().<br>
Secara garis besar Figure merupakan template dari objek visualisasinya sedangkan Axis merupakan sumbu dari Figure yang dibentuk dengan ketentuan yang telah ditetapkan.<br>

<h2>Soal 2: Membuat Component Figure dan Axis</h2>

- Buatlah Figure dan Axis kosong tanpa Data
- Plot data di bawah ini dengan 2 cara, yaitu dengan membuat figure dan axis secara explicit dan Implicit

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 20, 100)
y = np.cos(x/2)
```

```python
fig = plt.figure()
ax = fig.add_subplot()

plt.show()

#Explicit

fig, ax = plt.subplots()
ax.plot(x, y)

plt.show()

#Implicit

plt.plot(x, y)

plt.show()
```

<h2>Soal 3: Memasukan multiple data ke dalam satu axis</h2>

- Plot ketiga data berikut ke dalam satu axis, dan gunakan seaborn style
- save data tersebut dengan nama file bebas

```python
x = np.linspace(0, 20, 100)
y = np.cos(x/2)
y2 = np.sin(x)
y3 = np.sin(2*x)
```

```python
fig, ax = plt.subplots()
plt.style.use?
ax.plot(x,y)
ax.plot(x,y2)
ax.plot(x,y3)
plt.show()
fig.savefig('bebas.png')
```
