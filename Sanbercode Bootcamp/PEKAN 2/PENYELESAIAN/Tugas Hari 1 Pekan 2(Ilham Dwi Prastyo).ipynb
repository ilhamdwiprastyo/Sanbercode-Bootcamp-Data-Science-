<h2>Soal 3: DataFrame vs Series</h2>

- Apa perbedaan Series dengan list dan dictionary
- Apa perbedaan dataframe dan series
- Buatlah suatu data series

Jawab: 1) Perbedaan Series dengan list dan dictionary adalah Secara garis besar Series merupakan kombinasi dari list dan dictionary, dimana Series indexingnya bisa menggunakan slicing seperti list sedangkan dictionary tidak bisa slicing, serta Series indexingnya bisa menggunakan labeling sedangkan list tidak bisa, key dictionary tidak bisa di duplicate sedangkan di Series bisa

2) Perbedaan dataframe dan series adalah terletak pada besar dimensi , pada series memiliki 1 dimensi sedangkan dataframe memiliki dimensi, pada series terdiri dari index saja, dan isi datanya bernama values, sedangkan pada dataframe terdiri dari index dan collumns karena 2 dimensi. Dataframe juga bisa berupa gabungan dari Series, dapat dikatakan Series merupakan himpunan bagian dari Dataframe

<h2>Soal 6: Practice Deskriptif Statistik</h2>

- Gunakan suatu metode dari pandas untuk mengetahui deskriptif statistik dari suatu data
- Berapa nilai rata-rata dari column price
- Berapa nilai standar deviasi dari column depth
- Berapa nilai maximum dari column carat

```python
data_diamond.describe()
```

```python
# code here

import matplotlib.pyplot as plt

grafik_pemusatan_data_price = data_diamond.get('price')
mean= grafik_pemusatan_data_price.mean()
median = grafik_pemusatan_data_price.median()
modus = grafik_pemusatan_data_price.mode()

plt.figure(figsize=(15,5))
plt.title('Diagaram Pemusatan Data pada Columns Price')
plt.hist(grafik_pemusatan_data_price,bins=35,color='grey')
plt.axvline(mean,color='red',label='Mean')
plt.axvline(median,color='yellow',label='Median')
plt.axvline(modus[0],color='green',label='Modus')
plt.xlabel('price')
plt.ylabel('frekuensi')
plt.legend()
plt.show()


#nilai rata-rata dari column price
print("\nnilai rata-rata dari column price adalah",data_diamond.get('price').mean())

#nilai standar deviasi dari column depth
print("nilai standar deviasi dari column depth adalah",data_diamond.get('depth').std())

#nilai maximum dari column carat
print("nilai maximum dari column carat adalah", data_diamond.get('carat').max())
```
